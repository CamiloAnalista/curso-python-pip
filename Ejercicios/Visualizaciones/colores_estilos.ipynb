```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('dark_background')
```

## Matlab

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'y-')
ax.plot(x,x+3,'b.-')
ax.plot(x,x+4,'go:')
```

#pyplot

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(x,x+1,color="green", alpha=0.2, linewidth=12, linestyle=':', marker='o',markersize=20, markerfacecolor='orange')
ax.plot(x,x+2,color="blue")
ax.plot(x,x+3,color="#66FF66")
ax.plot(x,x+4,color="#CCFFCC")
```
