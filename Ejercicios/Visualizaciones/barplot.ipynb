```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pais = ['Japan','Germany','Columbia','Brazil']
poblacion = [100,200,700,400]
plt.bar(pais,poblacion,width=0.5, color=['red','blue','green','orange'])
plt.xticks(np.arange(4),('Japon','Alemania','Colombia','Brasil'),rotation=90)
plt.title('Población por país')
plt.ylabel('Población')
plt.xlabel('País')
plt.legend()
plt.show()
```
