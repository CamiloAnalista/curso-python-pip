```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
sns.set(style='darkgrid',palette='dark',font="Arial",font_scale=1)
sns.barplot(x=['A','B','C'],y=[1,3,2])
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.displot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='total_bill',y='tip',hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill',hue='sex')
plt.show()
```

```python
sns.displot(data=tips, x='total_bill',hue='sex',kind='kde')
plt.show()
```

```python
tips.head()
```

```python
sns.histplot(data=tips,x='tip',cumulative=False,hue='sex',stat='frequency',multiple='fill')
plt.show()
```

```python
sns.kdeplot(data=tips,x='tip',hue='sex',cumulative=False,shade=True)
plt.show()
```

```python
sns.ecdfplot(data=tips,x='tip',hue='sex',stat='count')
plt.show()
```

```python
sns.displot(data=tips,x='tip',hue='sex',kind='kde')
plt.show()
```
